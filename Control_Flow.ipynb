```python
age = input('Enter your age: ')

# convert the string to int
your_age = int(age)

# determine the ticket price
if your_age < 5:
  ticket_price = 5_000
elif your_age < 16:
  ticket_price = 10_000
else:
  ticket_price = 20_000

# show the ticket price
print(f"You'll pay {ticket_price} rupiah for the ticket")
```
