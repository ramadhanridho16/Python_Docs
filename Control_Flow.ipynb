# Python Ternary Operator

```python
# To make if else statement more concise u can use ternary operator

age = input('Enter your age: ')

if int(age) >= 18:
    ticket_price = 20
else:
    ticket_price = 5

print(f"The ticket price is {ticket_price}")
```

# Python While

## Introduction to the Python while statement

image.png

```python
# While loop must have a rule to stop or it will be indefinite loop that will run forever until close application

max = 5
counter = 0

while counter < max:
  print(counter)
  counter += 1
```

## Using Python while statement to build a simple command prompt program

```python
command = " "

while command.lower() != 'quit':
  command = input('>')
  print(f"Echo: {command}")
```

## Using Python break with for loop

image.png

```python
for index in range(0, 10):
  print(index)
  if index == 3:
    break
```

```python
# Using break statement in nested loop

for x in range(5):
  for y in range(5):
    # terminate the innermost loop
    if y > 1:
      break
    
    # shows cordinates on the screen
    print(f"({x}, {y})")
```

## Using Python break statement with a while loop

image.png

```python
# made a simple command prompt

print('-- Help: type quit to exit --')
while True:
  color = input('Enter your favorite color:')
  if color.lower() == 'quit':
    break
```
