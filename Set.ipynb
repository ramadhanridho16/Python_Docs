```python
# An empty set evaluates to False in Boolean context. For example:

skills = set()

if not skills:
  print(
      'Empty sets are falsy'
  )
```

```python
"""
In fact, you can pass an iterable to the set() function to create a set. 
For example, you can pass a list, which is an iterable, to the set() function like this:
"""

skills = set(['Problem solving', 'Critical Thingking'])
print(skills)
```

```python
"""
If an iterable has duplicate elements, 
the set() function will remove them. For example:
"""

characters = set('letter')
print(characters)
```

## Checking if an element is in a set

image.png

```python
ratings = {1, 2, 3, 4, 5}
rating = 1

if rating in ratings:
  print(f'The set contains {rating}')
```

```python
# To negate the in operator, you use the not operator. For example:

ratings = {1, 2, 3, 4, 5}
rating = 6

if rating not in ratings:
  print(f"The set does not contain {rating}")
```

## Looping through set elements

```python
skills = {'Problem solving', 'Software design', 'Python programming'}

for skill in skills:
  print(skill)
```

```python
# To access the index of the current element inside the loop, you can use the built-in enumerate() function:

skills = {'Problem solving', 'Software design', 'Python programming'}

for index, skill in enumerate(skills):
  print(f"{index}.{skill}")
```

```python
'''
By default, the index starts at zero. To change this, 
you pass the starting value to the second argument of the enumerate() function. For example:
'''

skills = {'Problem solving', 'Software design', 'Python programming'}
for index, skill in enumerate(skills, 1):
  print(f"{index}. {skill}")
```

# Set Comprehension

image.png

```python
# cara lama : for loop

tags = {'Django', 'Pandas', 'Numpy'}

lowercase_tags = set()
for tag in tags:
  lowercase_tags.add(tag.lower())

print(lowercase_tags)
```

```python
# Menggunakan map() function dengan lambda expression

tags = {'Django', 'Pandas', 'Numpy'}
lowercase_tags = set(map(lambda tag: tag.lower(), tags))

print(lowercase_tags)
```

## Set Comprehension Syntax

image.png

```python
# To make the code more concise, Python provides you with the set comprehension syntax as follows:

tags = {'Django', 'Pandas', 'Numpy'}

lowercase_tags = {tag.lower() for tag in tags}

print(lowercase_tags)
```

## Python Set comprehension with an if clause example

image.png

```python
tags = {'Django', 'Pandas', 'Numpy'}
new_tags = {tag.lower() for tag in tags if tag != 'Numpy'}

print(new_tags)
```
