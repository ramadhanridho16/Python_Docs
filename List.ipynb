## Modifying, adding, and removing elements

### 1) Modifying elements in a list

image.png

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers[0] = 10

print(numbers)
```

```python
# multiply element in list

numbers = [1, 3, 2, 7, 9, 4]
numbers[1] = numbers[1]*10

print(numbers)
```

```python
# divides the third element by 2:

numbers = [1, 3, 2, 7, 9, 4]
numbers[2] /= 2

print(numbers)
```

### 2) Adding elememts to the list

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers.append(100)

print(numbers)
```

```python
# insert() method adss a new element at any position in the list

numbers = [1, 3, 2, 7, 9, 4]
numbers.insert(2, 100)

print(numbers)
```

### 3) Removing elements from a list


```python
# del statement

numbers = [1, 3, 2, 7, 9, 4]
del numbers[0]

print(numbers)
```

```python
# The pop() method removes the last element from a list and returns that element:

numbers = [1, 3, 2, 7, 9, 4]
last = numbers.pop()

print(last)
print(numbers)
```

```python
'''
Typically, you use the pop() method when you want to remove an element from a list 
and still want to access the value of that element.

To pop an element by its position, you use the pop() with the element’s index. For example:
'''

numbers = [1, 3, 2, 7, 9, 4]

second = numbers.pop(1)

print(second)
print(numbers)
```

```python
# To remove an element by value, you use the remove() method.

numbers = [1, 3, 2, 7, 9, 4]

numbers.remove(9)
print(numbers)
```
