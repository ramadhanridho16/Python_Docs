## Modifying, adding, and removing elements

### 1) Modifying elements in a list

image.png

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers[0] = 10

print(numbers)
```

```python
# multiply element in list

numbers = [1, 3, 2, 7, 9, 4]
numbers[1] = numbers[1]*10

print(numbers)
```

```python
# divides the third element by 2:

numbers = [1, 3, 2, 7, 9, 4]
numbers[2] /= 2

print(numbers)
```

### 2) Adding elememts to the list

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers.append(100)

print(numbers)
```

```python
# insert() method adss a new element at any position in the list

numbers = [1, 3, 2, 7, 9, 4]
numbers.insert(2, 100)

print(numbers)
```

### 3) Removing elements from a list


```python
# del statement

numbers = [1, 3, 2, 7, 9, 4]
del numbers[0]

print(numbers)
```

```python
# The pop() method removes the last element from a list and returns that element:

numbers = [1, 3, 2, 7, 9, 4]
last = numbers.pop()

print(last)
print(numbers)
```

```python
'''
Typically, you use the pop() method when you want to remove an element from a list 
and still want to access the value of that element.

To pop an element by its position, you use the pop() with the elements index. For example:
'''

numbers = [1, 3, 2, 7, 9, 4]

second = numbers.pop(1)

print(second)
print(numbers)
```

```python
# To remove an element by value, you use the remove() method.

numbers = [1, 3, 2, 7, 9, 4]

numbers.remove(9)
print(numbers)
```

## Python List sort() method examples

### 1) Using the Python List sort() method to sort a list of strings


```python
# If a list contains strings, the sort() method sorts the string elements alphabetically.

guests = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']
guests.sort()

print(guests)
```

```python
''' And the following example uses the sort() method with the reverse=True 
argument to sort the elements in the guests list in the reverse alphabetical order:
'''

guests = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']
guests.sort(reverse=True)

print(guests)
```

### 2) Using the Python List sort() method to sort a list of numbers

```python
scores = [5, 7, 4, 6, 9, 8]
scores.sort()

print(scores)
```

```python
# largest to smallest

scores = [5, 7, 4, 6, 9, 8]
scores.sort(reverse=True)

print(scores)
```

### 3) Using the Python List sort() method to sort a list of tuples

image.png



```python
# Note that you just pass the function name without the parentheses to the sort() method:

companies = [('Google', 2019, 134.81),
             ('Apple', 2019, 260.2),
             ('Facebook', 2019, 70.7)]

# define a sort key
def sort_key(company):
  return company[2]

# sort the companies by revenue
companies.sort(key=sort_key, reverse=True)

# show the sorted companies
print(companies)
```

#### Using Lambda Expression

image.png

```python
# The following example uses the lambda expression to sort the companies by revenue from low to high:

companies = [('Google', 2019, 134.81),
             ('Apple', 2019, 260.2),
             ('Facebook', 2019, 70.7)]

# sort companies by revenue
companies.sort(key=lambda company:company[2])

# show the sorted companies
print(companies)
```

### 1) Using Python sorted() function to sort a list of strings

## Introduction

image.png

## Python List Slice Examples

### 1) Basic Python list slice example


```python
# example

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
sub_colors = colors[1:4]

print(sub_colors)

'''
This example doesnt use the step, 
so the slice gets all values within the range without skipping any elements.
'''
```

### 2) Using Python List slice to get the n-first elements from a list

image.png

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
sub_colors = colors[:4]

print(sub_colors)

# it's equivalent to the color[0:4]
```

### 3) Using Python List slice to get the n-last elements from a list

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
sub_colors = colors[-3:]

print(sub_colors)
```

### 4) Using Python List slice to get every nth-element from a list


```python
# uses the step to return a sublist that includes every 2nd element of the color list

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
sub_colors = colors[::2]

print(sub_colors)
```

### 5) Using Python List slice to reverse a list


```python
# last element to first element (reverse)

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
reversed_colors = colors[::-1]

print(reversed_colors)
```

### 6) Using Python List slice to substitute part a list

image.png

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
colors[0:2] = ['black', 'white']

print(colors)
```

### 7) Using python List slice to partially replace and resize a list

image.png

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
print(f"The list has {len(colors)} elements")

colors[0:2] = ['black', 'white', 'gray']
print(colors)

print(f"\nThe list has {len(colors)} elements")
```

### 8) Using Python list slice to delete elements


```python
'''
The following shows how to use the list slice 
to delete the 3rd, 4th, and 5th elements from the colors list:
'''

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
del colors[2:5]

print(colors)
```
