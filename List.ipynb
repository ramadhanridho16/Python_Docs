## Modifying, adding, and removing elements

### 1) Modifying elements in a list

image.png

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers[0] = 10

print(numbers)
```

```python
# multiply element in list

numbers = [1, 3, 2, 7, 9, 4]
numbers[1] = numbers[1]*10

print(numbers)
```

```python
# divides the third element by 2:

numbers = [1, 3, 2, 7, 9, 4]
numbers[2] /= 2

print(numbers)
```

### 2) Adding elememts to the list

```python
numbers = [1, 3, 2, 7, 9, 4]
numbers.append(100)

print(numbers)
```

```python
# insert() method adss a new element at any position in the list

numbers = [1, 3, 2, 7, 9, 4]
numbers.insert(2, 100)

print(numbers)
```

### 3) Removing elements from a list


```python
# del statement

numbers = [1, 3, 2, 7, 9, 4]
del numbers[0]

print(numbers)
```

```python
# The pop() method removes the last element from a list and returns that element:

numbers = [1, 3, 2, 7, 9, 4]
last = numbers.pop()

print(last)
print(numbers)
```

```python
'''
Typically, you use the pop() method when you want to remove an element from a list 
and still want to access the value of that element.

To pop an element by its position, you use the pop() with the elements index. For example:
'''

numbers = [1, 3, 2, 7, 9, 4]

second = numbers.pop(1)

print(second)
print(numbers)
```

```python
# To remove an element by value, you use the remove() method.

numbers = [1, 3, 2, 7, 9, 4]

numbers.remove(9)
print(numbers)
```

```python
# Once you define a tuple, you can access an individual element by its index, just. For example:

rgb = ('red', 'green', 'blue')

print(rgb[0])
print(rgb[1])
print(rgb[2])
```

## Python List sort() method examples

### 1) Using the Python List sort() method to sort a list of strings


```python
# If a list contains strings, the sort() method sorts the string elements alphabetically.

guests = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']
guests.sort()

print(guests)
```

```python
''' And the following example uses the sort() method with the reverse=True 
argument to sort the elements in the guests list in the reverse alphabetical order:
'''

guests = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer']
guests.sort(reverse=True)

print(guests)
```

### 2) Using the Python List sort() method to sort a list of numbers

```python
scores = [5, 7, 4, 6, 9, 8]
scores.sort()

print(scores)
```

```python
# largest to smallest

scores = [5, 7, 4, 6, 9, 8]
scores.sort(reverse=True)

print(scores)
```

### 3) Using the Python List sort() method to sort a list of tuples

image.png



```python
# Note that you just pass the function name without the parentheses to the sort() method:

companies = [('Google', 2019, 134.81),
             ('Apple', 2019, 260.2),
             ('Facebook', 2019, 70.7)]

# define a sort key
def sort_key(company):
  return company[2]

# sort the companies by revenue
companies.sort(key=sort_key, reverse=True)

# show the sorted companies
print(companies)
```

#### Using Lambda Expression

image.png

```python
# The following example uses the lambda expression to sort the companies by revenue from low to high:

companies = [('Google', 2019, 134.81),
             ('Apple', 2019, 260.2),
             ('Facebook', 2019, 70.7)]

# sort companies by revenue
companies.sort(key=lambda company:company[2])

# show the sorted companies
print(companies)
```
