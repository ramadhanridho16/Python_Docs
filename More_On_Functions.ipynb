## Using Unpacking tuple to swap values of two variables

### Using Logic Way

```python
# Tuple is mutable but it can be swapped

x = 10
y = 20

print(f"x = {x}, y = {y}")

tmp = x
x = y
y = tmp

print(f"x = {x}, y = {y}")
```

### Using unpacking tuple way

```python
x = 10
y = 20

print(f"x={x}, y={y}")

x, y = y, x

print(f"x={x}, y={y}")
```

## Tuple unpacking

```python
def add(x, y, *args):
  total = x + y
  for arg in args:
    total += arg

  return total
```

```python
result = add(10, 20, 30, 40, 50)
print(result)
```

## Python *args argument exhaust positional arguments

```python
# !False argument (syntax error)

def add(x, y, *args, z):
  return x + y + sum(args) + z

# False 
# add(10, 20, 30, 40, 50)

# True
add(10, 20, 30, 40, z=50)
```

# Python Partial Functions

image.png

```python
# Partial Function (Functions that multiplies two arguments)

def multiply(a, b):
  return a*b
```

```python
def double(a):
  return multiply(a, 2)
```

```python
double(10)
```

image.png

## Python partial function from functools module

image.png

```python
from functools import partial

def multiply(a, b):
  return a*b
```

```python
help(partial)
```

```python
doule = partial(multiply, b=2)
result = double(10)
print(result)
```

image.png

## Python partial functions and variables

```python
# first, running the multiply func at statement above

x = 2
f = partial(multiply, x)
```

```python
result = f(10)
result
```

```python
x = 3
result = f(10)
print(result)
```

## Adding type hints for multiple types

```python
def add(x, y):
  return x+y
```

image.png

```python
from typing import Union
```

```python
# Second, use the Union to create a union type that includes int and float:

def add(x: Union[int, float], y: Union[int, float]) -> Union[int, float]:
  return x+y
```

```python
# Starting from Python 3.10, you can use the X | Y syntax to create a union type, for example:

# def tambah(x: int|float, y:int|float) -> int|float:
  # return x + y
```

## Type aliases

image.png

```python
from typing import Union

number = Union[int, float]

def add(x: number, y: number) -> number:
  return x + y
```
