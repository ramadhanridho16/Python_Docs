## Using Unpacking tuple to swap values of two variables

### Using Logic Way

```python
# Tuple is mutable but it can be swapped

x = 10
y = 20

print(f"x = {x}, y = {y}")

tmp = x
x = y
y = tmp

print(f"x = {x}, y = {y}")
```

### Using unpacking tuple way

```python
x = 10
y = 20

print(f"x={x}, y={y}")

x, y = y, x

print(f"x={x}, y={y}")
```

## Using the * operator on the right hand side

```python
odd_numbers = (1, 3, 5)
even_numbers = (2, 4, 6)

# merge into a single tuple
numbers = (*odd_numbers, *even_numbers)
print(numbers)
```

## Tuple unpacking

```python
def add(x, y, *args):
  total = x + y
  for arg in args:
    total += arg

  return total
```

```python
result = add(10, 20, 30, 40, 50)
print(result)
```

## Introduction to the Python *args parameter

image.png

```python
def add(*args):
  print(args)

add()
```

```python
# The add function shows an empty tuple

def add(*args):
  print(type(args))
  print(args)

add()
```

```python
def add(*args):
  print(args[0])
  print(args[1])
  print(args[2])

add(1, 2, 3)
```

image.png

```python
def add(*args):
  total = 0
  for arg in args:
    total += arg
  return total

add(1, 2, 3)
```

## Python *args argument exhaust positional arguments

image.png

```python
# !False argument (syntax error)

def add(x, y, *args, z):
  return x + y + sum(args) + z

# False 
# add(10, 20, 30, 40, 50)

# True
add(10, 20, 30, 40, z=50)
```
