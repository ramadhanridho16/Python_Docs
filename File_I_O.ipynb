```python
with open('/content/drive/MyDrive/3. Colab Notebooks/Python Documentation/File/ridho.txt') as data:
  while True:
    line = data.readline()
    if not line:
      break
    print(line.strip())
```

## Concise way to read a text file line by line

image.png

```python
with open('/content/drive/MyDrive/3. Colab Notebooks/Python Documentation/File/ridho.txt') as f:
  for line in f:
    print(line.strip())
```
