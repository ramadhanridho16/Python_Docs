## Keyword arguments and default parameters

```python
# fuction that calculates net price from selling price, tax, and discount

def get_net_price(price, tax=0.07, discount = 0.05):
  return price * (1 + tax - discount)

net_price = get_net_price(100)
print(net_price)
```

```python
# using keyword arguments and default parameters

net_price = get_net_price(100, discount=0.06)
print(net_price)
```

### 2) Using a recursive function to calculate the sum of a sequence

image.png

```python
def sum(n):
  total = 0
  for index in range(n+1):
    total += index
    print(f"{index} + ")

  return total
```

```python
result = sum(5)
print(f"= {result}")
```

```python
# using ternary operator
# value_if_true "if" conditional_ "else" not_true/false

def sum(n):
  return n + sum(n-1) if n > 0 else 0

sum(5)    # why this program can work after remove print function???
```

## Python lambda expression examples

In Python, you can pass a func to another func or return a func from another func

### 1) Functions that accept a function example

```python
def get_full_name(first_name, last_name, formatter):
  return formatter(first_name, last_name)

def first_last(first_name, last_name):
  return f"{first_name} {last_name}"

def last_first(first_name, last_name):
  return f"{last_name}, {first_name}"
```

```python
full_name = get_full_name('John', 'Doe', first_last)
print(full_name)
```

```python
full_name = get_full_name('John', 'Doe', last_first)
print(full_name)
```

image.png

```python
# Alternate function statement in above with lambda expressions

def get_full_name(first_name, last_name, formatter):
  return formatter(first_name, last_name)

full_name = get_full_name(
    'John',
    'Doe',
    lambda first_name, last_name: f"{first_name} {last_name}"
)
print(full_name)
```

```python
full_name_2 = get_full_name(
    'John',
    'Doe',
    lambda first_name, last_name: f"{last_name} {first_name}"
)
print(full_name_2)
```

### 2) Functions that return a function example

```python
# The following times() function returns a function which is a lambda expression:

def times(n):
  return lambda x: x*n
```

```python
double = times(2)
```

```python
# Since the times() functions returns a function, the double is also a function. To call it, you place parethesses like this:
result = double(2)
print(result)
```

```python
result = double(3)
print(result)
```

```python
triple = times(3)

print(triple(2))
print(triple(3))
```

## Python lambda in a loop

```python
callables = []
for i in (1, 2, 3):
  callables.append(lambda: i)

for f in callables:
  print(f())
```
