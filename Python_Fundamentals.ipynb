# Python String

## 1. Introduction to Python String

```python
message = 'This is a string in Python'
message = "This is also a string"

# If a string contains single quote, you should place it in double-quotes like this.
message = "It's a string"
print(message)
```

```python
# When a string contains double quotes, you can use the single quotes

message = '"Beautiful is better than ugly.". Said Tim Peters'
print(message)
```

```python
# The python interpreter will treat backslash character special. If you don't want to do it, u can se raw strings by addin letter 'r'

message = r'C:\python\bin'
print(message)
```

## 2. Creating multiline strings

```python
help_message = '''
Usage: mysql command
  -h hostname
  -d database name
  -u username
  -p password
'''
print(help_message)
```

## 3. Using Variables in Python strings with the f-strings

```python
name = 'John'
message = 'Hi'

# To do it, you place the letter f before the opening quotation mark and put the brace around the variable name:
name = 'Ramadhan'
message = f'Hi {name}'
print(message)
```

## 4. Concatenation Python Strings

```python
greeting = 'Good ' 'Morning!'
print(greeting)
```

```python
# concatenate two string variables, using operator '+'

greeting = 'Good'
time = 'Afternoon'

greeting = greeting + time + '!'
print(greeting)
```

## 7. Slicing strings

```python
str = "Python String"
print(str[0:2])     # it will be stop before index 2
```

```python
# The Syntax
# string [start:end]
# Python string are immutable

str = "Python String"
str[0] = 'J'
```

```python
# When want to modify string, must create a new one from existing string

str = "Python String"
new_str = 'J' + str[1:]
print(new_str)
```
